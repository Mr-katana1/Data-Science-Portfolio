```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python

match_data = {
    'Match_ID': [201, 202, 203, 204, 205, 206, 207],
    'Game_Mode': ['Solo', 'solo ', 'DUO', 'Squads', 'duo', 'Solo', 'Squads '],
    'Kills': [3, '5 kills', 0, 'none', 2, 4, '1 kill'],
    'Damage_Dealt': ['1,200', '1,550', '250', '800', '1,100', None, '450'],
    'Placement': [5, 1, 22, 8, 3, 12, 15]
}

df_gaming = pd.DataFrame(match_data)
print("Esports Dataset Loaded! Total matches:", len(df_gaming))
```

```python
df_gaming.Game_Mode=df_gaming.Game_Mode.str.strip().str.lower()
```

```python
df_gaming.Kills = df_gaming.Kills.astype(str).str.extract(r'(\d+)').fillna(0).astype(int)
```

```python
df_gaming.Damage_Dealt=df_gaming.Damage_Dealt.str.replace(',','').fillna(850).astype(float)
```

```python
df_gaming.head()
```

```python
df_gaming.groupby('Game_Mode')[['Damage_Dealt','Placement']].mean()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(df_gaming,x='Placement',y='Damage_Dealt',hue='Game_Mode')
plt.title('Placement VS Damage Dealt Across Diffrent Gamemode', fontsize=14, fontweight='bold')
plt.grid(True, alpha=0.3)
plt.show()
```
